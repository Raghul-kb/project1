# Task 2: Model Implementation

```python
# Import required libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
# Separate features and target
X = df.drop('letter', axis=1)
y = df['letter']
```

```python
# Encode the target labels (alphabets)
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
# Normalize feature data
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Split dataset into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)
```

```python
# One-hot encode the labels for multi-class classification
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
# Build the ANN model (Default Model)
model_default = Sequential([
    Dense(64, input_dim=X_train.shape[1], activation='relu'),  # Hidden Layer 1
    Dense(32, activation='relu'),                             # Hidden Layer 2
    Dense(y_train_categorical.shape[1], activation='softmax')  # Output Layer
])
```

```python
#  Compile the model
model_default.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
#  Train the model
history = model_default.fit(
    X_train, y_train_categorical,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
#  Evaluate the model on test data
loss, accuracy = model_default.evaluate(X_test, y_test_categorical, verbose=0)
print("\nâœ… Model Evaluation:")
print(f"Test Loss: {loss:.4f}")
print(f"Test Accuracy: {accuracy:.4f}")
```

```python
#  Make predictions
y_pred_prob = model_default.predict(X_test)
y_pred_classes = np.argmax(y_pred_prob, axis=1)
```

```python
print("\nSample Predictions (first 10):")
print("Predicted:", label_encoder.inverse_transform(y_pred_classes[:10]))
print("Actual:", label_encoder.inverse_transform(y_test[:10]))
```

# Task 3: Hyperparameter Tuning

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import RandomizedSearchCV
from scikeras.wrappers import KerasClassifier
import numpy as np
```

```python
# Function to create the ANN model
def create_model(learning_rate=0.001, neurons=64, activation='relu', dropout_rate=0.0):
    model = Sequential()
    model.add(Dense(neurons, input_dim=X_train.shape[1], activation=activation))
    model.add(Dropout(dropout_rate))
    model.add(Dense(neurons // 2, activation=activation))
    model.add(Dense(len(np.unique(y_train)), activation='softmax'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Wrap model for use in GridSearchCV
model = KerasClassifier(model=create_model, verbose=0)
```

```python
# Hyperparameter grid
param_grid = {
    'model__neurons': [64],
    'model__activation': ['relu', 'tanh'],
    'model__dropout_rate': [0.0],
    'model__learning_rate': [0.001, 0.01],
    'batch_size': [32],
    'epochs': [40]
}
```

```python
# Perform grid search (uses 3-fold cross-validation)
grid = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_iter=10, cv=3, verbose=1, n_jobs=-1)
grid_result = grid.fit(X_train, y_train)
```

```python
# Display the best hyperparameters and accuracy
print("\nâœ… Best Hyperparameters Found:")
print(grid_result.best_params_)
print("\nBest Cross-Validation Accuracy: {:.4f}".format(grid_result.best_score_))
```

```python
# Evaluate the tuned model on the test data
best_model = grid_result.best_estimator_.model_
test_loss, test_acc = best_model.evaluate(X_test, y_test, verbose=0)
print("\nâœ… Tuned Model Evaluation on Test Set:")
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_acc:.4f}")
```
